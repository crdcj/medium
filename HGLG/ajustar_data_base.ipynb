```python
import pandas as pd
```

```python
# Histórico de preços de fechamento do HGLG11
hglg_preult = pd.read_csv('data/hglg_preult.csv', parse_dates=['datneg'])
hglg_preult
```

```python
# Histórico de dividendos do HGLG11
hglg_div = (
    pd.read_csv('data/hglg_div.csv', parse_dates=['data_base'])
    [['data_base', 'valor_por_cota']]
    .sort_values('data_base', ignore_index=True)
)
hglg_div
```

```python
# Ajustar as datas de dividendos para que a data tenha tido necessariamente negociação
hglg_div['data_base_ajustada'] = hglg_div['data_base'].apply(
    lambda x: hglg_preult.query('datneg <= @x')['datneg'].max()
)
hglg_div
```

```python
hglg_div.query('data_base != data_base_ajustada')
```

```python

# Verificar quais datas foram ajustadas
hglg_div.query('data_base != data_base_ajustada')
```

```python
# Remover a coluna data_base
hglg_div.drop('data_base', axis=1, inplace=True)
# Renomear a coluna data_base_ajustada para data_base
hglg_div.rename(columns={'data_base_ajustada': 'data_base'}, inplace=True)
hglg_div
```

```python
# Salvar o arquivo
hglg_div.to_csv('data/hglg_div_data_base_ajustada.csv', index=False)
```
